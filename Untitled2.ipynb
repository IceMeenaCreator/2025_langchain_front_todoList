```python



# ------------------------------------------------------------
# 1. Helper functions
# ------------------------------------------------------------

def norm_biz(x):
    if pd.isna(x):
        return ""
    digits = re.sub(r"\D", "", str(x))
    return digits if len(digits) == 10 else ""


def to_num(x):
    return pd.to_numeric(x, errors="coerce")


def to_date(x):
    return pd.to_datetime(x, errors="coerce")


def safe_divide(a, b):
    a = pd.to_numeric(a, errors="coerce")
    b = pd.to_numeric(b, errors="coerce")
    return (a / b.replace({0: np.nan})).replace([np.inf, -np.inf], np.nan)


def winsorize(s, lower=0.01, upper=0.99):
    s = pd.to_numeric(s, errors="coerce").replace([np.inf, -np.inf], np.nan)
    if s.notna().sum() < 30:
        return s
    lo, hi = s.quantile([lower, upper])
    return s.clip(lo, hi)


def first_nonmissing(s):
    for v in s:
        if pd.notna(v) and str(v).strip() != "":
            return v
    return np.nan


def smd(treated, control):
    treated = pd.to_numeric(treated, errors="coerce").dropna()
    control = pd.to_numeric(control, errors="coerce").dropna()
    if len(treated) < 2 or len(control) < 2:
        return np.nan
    pooled = math.sqrt((treated.var(ddof=1) + control.var(ddof=1)) / 2)
    if pooled == 0:
        return 0.0
    return float((treated.mean() - control.mean()) / pooled)


def stars(p):
    if pd.isna(p):
        return ""
    if p < 0.01:
        return "***"
    if p < 0.05:
        return "**"
    if p < 0.10:
        return "*"
    return ""


def summary_table(df, group_cols, vars_):
    rows = []
    for keys, g in df.groupby(group_cols, dropna=False):
        if not isinstance(keys, tuple):
            keys = (keys,)
        row = dict(zip(group_cols, keys))
        row["n_obs"] = len(g)
        if "firm_id" in g.columns:
            row["n_firms"] = g["firm_id"].nunique()
        if "loan_event_id" in g.columns:
            row["n_loan_events"] = g["loan_event_id"].nunique()
        for v in vars_:
            if v not in g.columns:
                continue
            s = pd.to_numeric(g[v], errors="coerce")
            row[f"{v}_n"] = int(s.notna().sum())
            row[f"{v}_mean"] = s.mean()
            row[f"{v}_std"] = s.std()
            row[f"{v}_min"] = s.min()
            row[f"{v}_p25"] = s.quantile(0.25)
            row[f"{v}_median"] = s.median()
            row[f"{v}_p75"] = s.quantile(0.75)
            row[f"{v}_max"] = s.max()
        rows.append(row)
    return pd.DataFrame(rows)
```

```python
def clean_cols(df):
    df = df.copy()
    df.columns = [str(c).strip().replace("\n", " ").replace("\t", " ") for c in df.columns]
    return df


def norm_biz(x):
    if pd.isna(x):
        return ""
    digits = re.sub(r"\D", "", str(x))
    return digits if len(digits) == 10 else ""


def to_num(x):
    return pd.to_numeric(x, errors="coerce")


def to_date(x):
    return pd.to_datetime(x, errors="coerce")


def stars(p):
    if pd.isna(p):
        return ""
    if p < 0.01:
        return "***"
    if p < 0.05:
        return "**"
    if p < 0.10:
        return "*"
    return ""


def safe_divide(a, b):
    a = pd.to_numeric(a, errors="coerce")
    b = pd.to_numeric(b, errors="coerce")
    return (a / b.replace({0: np.nan})).replace([np.inf, -np.inf], np.nan)


def summary_stats(df, group_cols, value_cols):
    rows = []
    for keys, g in df.groupby(group_cols, dropna=False):
        if not isinstance(keys, tuple):
            keys = (keys,)
        row = dict(zip(group_cols, keys))
        row["n_obs"] = len(g)
        if "firm_id" in g.columns:
            row["n_firms"] = g["firm_id"].nunique()
        if "loan_event_id" in g.columns:
            row["n_loan_events"] = g["loan_event_id"].nunique()

        for v in value_cols:
            if v not in g.columns:
                continue
            s = pd.to_numeric(g[v], errors="coerce")
            row[f"{v}_n"] = int(s.notna().sum())
            row[f"{v}_mean"] = s.mean()
            row[f"{v}_std"] = s.std()
            row[f"{v}_min"] = s.min()
            row[f"{v}_p25"] = s.quantile(0.25)
            row[f"{v}_median"] = s.median()
            row[f"{v}_p75"] = s.quantile(0.75)
            row[f"{v}_max"] = s.max()
        rows.append(row)

    return pd.DataFrame(rows)


def smd(t, c):
    t = pd.to_numeric(t, errors="coerce").dropna()
    c = pd.to_numeric(c, errors="coerce").dropna()
    if len(t) < 2 or len(c) < 2:
        return np.nan
    pooled = np.sqrt((t.var(ddof=1) + c.var(ddof=1)) / 2)
    if pooled == 0:
        return 0
    return (t.mean() - c.mean()) / pooled
```

```python
def standardize_panel(df):
    df = df.copy()

    if "firm_id" not in df.columns:
        if "biz_number_clean" in df.columns:
            df["firm_id"] = df["biz_number_clean"].map(norm_biz)
        elif "biz_number" in df.columns:
            df["firm_id"] = df["biz_number"].map(norm_biz)

    for c in ["panel_year", "year", "financial_year", "source_year"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    if "panel_year" not in df.columns:
        for c in ["year", "financial_year", "source_year"]:
            if c in df.columns:
                df["panel_year"] = df[c]
                break

    if "relative_year" in df.columns:
        df["relative_year"] = pd.to_numeric(df["relative_year"], errors="coerce")

    if "first_loan_year" in df.columns:
        df["first_loan_year"] = pd.to_numeric(df["first_loan_year"], errors="coerce")

    if "loan_execution_year_int" in df.columns:
        df["loan_execution_year_int"] = pd.to_numeric(df["loan_execution_year_int"], errors="coerce")

    if "assigned_t0" in df.columns:
        df["assigned_t0"] = pd.to_numeric(df["assigned_t0"], errors="coerce")

    if "treated" in df.columns:
        df["treated"] = pd.to_numeric(df["treated"], errors="coerce").fillna(0).astype(int)

    return df


for name in list(data.keys()):
    data[name] = standardize_panel(data[name])

loan_event = data.get("loan_event")
firm_master = data.get("firm_master")
treated_event = data.get("treated_event")
balanced_core = data.get("balanced_core")
balanced_debt = data.get("balanced_debt")
status_df = data.get("status")
control = data.get("control")
matched_panel = data.get("matched_panel")
treated_baseline = data.get("treated_baseline")
```

# **14. DID / event-study 함수**

```python
def run_fe_ols(df, yvar, xvars, fe_type="firm_year", cluster="firm_id"):
    """
    fe_type:
    - firm_year: firm/panel-unit FE + year FE
    - firm_industry_year: firm/panel-unit FE + industry-year FE
    """
    d = df.copy()
    d = d.dropna(subset=[yvar] + xvars + ["panel_unit_id", "panel_year", cluster])

    if len(d) < 50 or d[yvar].nunique(dropna=True) < 2:
        return pd.DataFrame([{
            "outcome": yvar,
            "term": "",
            "coef": np.nan,
            "se": np.nan,
            "t": np.nan,
            "p_value": np.nan,
            "stars": "",
            "nobs": len(d),
            "fe_type": fe_type,
            "error": "insufficient data"
        }])

    rhs = " + ".join(xvars)

    if fe_type == "firm_year":
        formula = f"{yvar} ~ {rhs} + C(panel_unit_id) + C(panel_year)"
    elif fe_type == "firm_industry_year":
        formula = f"{yvar} ~ {rhs} + C(panel_unit_id) + C(industry_year)"
    else:
        raise ValueError("Unknown fe_type")

    model = smf.ols(formula, data=d)
    res = model.fit(cov_type="cluster", cov_kwds={"groups": d[cluster].astype(str)})

    rows = []
    conf = res.conf_int()

    for x in xvars:
        p = res.pvalues.get(x, np.nan)
        rows.append({
            "outcome": yvar,
            "term": x,
            "coef": res.params.get(x, np.nan),
            "se": res.bse.get(x, np.nan),
            "t": res.tvalues.get(x, np.nan),
            "p_value": p,
            "stars": stars(p),
            "ci95_low": conf.loc[x, 0] if x in conf.index else np.nan,
            "ci95_high": conf.loc[x, 1] if x in conf.index else np.nan,
            "nobs": int(res.nobs),
            "n_firms": d["firm_id"].nunique(),
            "fe_type": fe_type
        })

    return pd.DataFrame(rows)


def run_many(df, outcomes, xvars, fe_type="firm_year"):
    out = []
    for y in outcomes:
        if y in df.columns:
            out.append(run_fe_ols(df, y, xvars, fe_type=fe_type))
    return pd.concat(out, ignore_index=True) if out else pd.DataFrame()
```

26. modern DID: Python식 cohort-specific ATT


아래 코드는 Callaway–Sant’Anna의 핵심 아이디어처럼 cohort별로 t−1 대비 변화량을 treated-control 간 비교합니다. 엄밀한 did 패키지의 doubly robust estimator는 R에서 수행하는 것이 좋지만, 이 Python 코드는 현재 matched panel에 바로 적용할 수 있는 cohort-time ATT입니다.

```python
def cohort_att_gt(df, outcome, base_rel=-1, rel_years=[0, 1, 2, 3]):
    rows = []

    d = df.copy()
    d = d.dropna(subset=[outcome, "cohort", "relative_year", "treated", "firm_id"])

    for g in sorted(d["cohort"].dropna().unique()):
        dg = d[d["cohort"].eq(g)].copy()

        base = dg[dg["relative_year"].eq(base_rel)][["firm_id", "treated", outcome]].rename(
            columns={outcome: f"{outcome}_base"}
        )

        dg = dg.merge(base[["firm_id", f"{outcome}_base"]], on="firm_id", how="left")
        dg[f"{outcome}_change"] = dg[outcome] - dg[f"{outcome}_base"]

        for t in rel_years:
            dt = dg[dg["relative_year"].eq(t)].copy()
            tr = dt[dt["treated"].eq(1)][f"{outcome}_change"].dropna()
            co = dt[dt["treated"].eq(0)][f"{outcome}_change"].dropna()

            if len(tr) < 5 or len(co) < 5:
                continue

            att = tr.mean() - co.mean()

            # simple Welch SE
            se = np.sqrt(tr.var(ddof=1) / len(tr) + co.var(ddof=1) / len(co))
            tval = att / se if se > 0 else np.nan
            pval = 2 * (1 - stats.norm.cdf(abs(tval))) if pd.notna(tval) else np.nan

            rows.append({
                "outcome": outcome,
                "cohort": g,
                "relative_year": t,
                "ATT_gt": att,
                "se": se,
                "t": tval,
                "p_value": pval,
                "stars": stars(pval),
                "n_treated": len(tr),
                "n_control": len(co)
            })

    return pd.DataFrame(rows)


modern_did_rows = []
for y in main_outcomes:
    modern_did_rows.append(cohort_att_gt(mp2, y))

modern_did_attgt = pd.concat(modern_did_rows, ignore_index=True)

display(modern_did_attgt)
modern_did_attgt.to_excel(OUT / "R03_modern_DID_cohort_ATTgt_python.xlsx", index=False)
```

```python
# ============================================================
# Sun and Abraham event-study using fixest
# ============================================================

library(fixest)
library(readr)
library(dplyr)

df <- read_csv("08_matched_psm_did_panel.csv")

df <- df %>%
  mutate(
    firm_numeric_id = as.numeric(as.factor(firm_id)),
    year = as.numeric(panel_year),
    cohort_sa = ifelse(treated == 1, as.numeric(cohort), NA),
    rel_year = as.numeric(relative_year)
  )

outcomes <- c(
  "log_assets",
  "log_sales",
  "debt_ratio",
  "short_term_debt_assets",
  "long_term_debt_assets",
  "roa",
  "roe",
  "op_profit_assets",
  "rd_asset_ratio",
  "intangible_ratio"
)

for (y in outcomes) {
  if (!(y %in% names(df))) next

  fml <- as.formula(
    paste0(y, " ~ sunab(cohort_sa, year, ref.p = -1) | firm_numeric_id + year")
  )

  m <- feols(
    fml,
    data = df,
    cluster = ~firm_numeric_id
  )

  print(summary(m))

  out <- broom::tidy(m)
  write.csv(out, paste0("sun_abraham_", y, ".csv"), row.names = FALSE)
}
```
